#**Rock, Paper and scissor game**
## Single player versus Computer

Rock Paper Scissors (also known by other orderings of the three items, with "rock" sometimes being called "stone", or as Rochambeau, roshambo, or ro-sham-bo) is a hand game originating from China, usually played between two people, in which each player simultaneously forms one of three shapes with an outstretched hand. These shapes are "rock" (a closed fist), "paper" (a flat hand), and "scissors" (a fist with the index finger and middle finger extended, forming a V). "

Pseudo code 

1. Creating choices = Rock , paper, scissor
2. Player selecting choice in the given list of choice
3. Computer selection random choice in the given list of choice 
4. Comparing and Printing the choice selected by computer and player and result of  win/lose/tie conditions of player against computer 


```python
# Step 1 : creating choices  
choices = ('rock', 'paper', 'scissors')


print(' Welcome to Rock, paper and scissor game \n')
print("Game Instructions : \n Player has to type his choice from [rock , paper , scissors] where capital words are allowed but not with different spelling \n")
print('About game : \n 1. Rock beats Scissors. \n 2. Scissors beat Paper \n 3. Paper beats Rocks   \n')
print('Lets start the game \n')

next_game = 'yes'

while next_game == 'yes': # creating a while for next game 


# step 2  : Player selecting choice in the given list of choice 
  player = None 

  while player not in choices:
    player = input('choose one in rock, paper , scissors :').lower()



  # step 3 : computer selecting random choice out of choices list
  ''' for random selection we have to import random module 
  reference : https://pynative.com/python-random-sample/
  In this case we can use either random.sample() or random.choices() method as the list of choices we are considering are unique ,
  hence with repetition or without repetition doesn't make any sense .'''
  # storing the random choice in computer variable


  import random
  computer = random.choices(choices)
  computer = computer[0] # after selecting random choice , the choice is stored in computer in the form of list , but for our comparision we wanted it in string so extracting element from the list



# step 4 : Comparing and Printing the choice selected by computer and player and result of win/lose/tie conditions of player against computer

  if player ==  computer:
    print('Player : ', player)
    print('Computer:' ,computer)
    print('\n')
    print(' There is tie !!')
    print('\n')

  elif computer == 'rock':
    if player == 'paper':
      print('Player : ', player)
      print('Computer:' ,computer)
      print('\n')
      print(' Congragulations, You win!! ')
      print('\n')


    elif player == 'scissors':
      print('Player : ', player)
      print('Computer:' ,computer)
      print('\n')
      print(' Better luck next time , You Lose! ')
      print('\n')

  elif computer == 'paper':
    if player == 'scissors':
      print('Player : ', player)
      print('Computer:' ,computer)
      print('\n')
      print(' Congragulations, You win!! ')
      print('\n')

    elif player == 'rock':
      print('Player : ', player)
      print('Computer:' ,computer)
      print('\n')
      print(' Better luck next time , You Lose! ')
      print('\n')

  elif computer == 'scissors':
    if player == 'rock':
      print('Player : ', player)
      print('Computer:' ,computer)
      print('\n')
      print(' Congragulations, You win!! ')
      print('\n')


    elif player == 'paper':
      print('Player : ', player)
      print('Computer:' ,computer)
      print('\n')
      print(' Better luck next time , You Lose! ')
      print('\n')

  # getting user input for next game
  next_game_choice  = input("Want to play one more game : Type 'y' for Yes and 'n' for No : \n (Enter a valid input else game will be terminated)").lower()  
  print('\n')


  if next_game_choice == 'n':
    next_game = 'no'
    print('Thanks for playing come back soon')
  elif next_game_choice == 'y':
    next_game = 'yes'
  else:
    next_game = 'no'
    print('Game is terminated , because of invalid input ')
    print('Thanks for playing come back soon')
```
